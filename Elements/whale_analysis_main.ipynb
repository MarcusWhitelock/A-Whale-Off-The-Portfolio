## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices into dataframe variable
sp_tsx_history_df = pd.read_csv(sp_tsx_history_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
# Sorting the index to ascending 
sp_tsx_history_df.sort_index(ascending=True, inplace=True)
# Printing to confirm everything is correct 
sp_tsx_history_df.head()
```

```python
# Checking the Data Types
sp_tsx_history_df.dtypes
```

```python
# Fixing the Data Types of the column 'close'
# Have values in the close column that are not numerals they are symbols and will replace them and then change column to float
sp_tsx_history_df['Close'] = sp_tsx_history_df['Close'].str.replace(',', '').str.replace('$', '').astype(float)
# calling it to confirm above changes worked 
sp_tsx_history_df.dtypes
```

```python
# Calculating the Daily Returns with pct_change 

sp_tsx_history_daily_returns = sp_tsx_history_df.pct_change()

#Calling to check if it worked

sp_tsx_history_daily_returns.head()
```

```python
# Droping the nulls
sp_tsx_history_daily_returns.dropna(inplace=True)
# Calling to confirm dropping worked
sp_tsx_history_daily_returns.head()
```

```python
# Renaming `Close` Column to 'S&P TSX 60 Returns' so we can see it with correct name when we concat.
sp_tsx_history_daily_returns.columns = ["S&P TSX 60 Returns"]

#Calling to view if rename worked
sp_tsx_history_daily_returns.head()
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios.
4. Determine which portfolios are riskier than the S&P TSX 60.
5. Calculate the Annualized Standard Deviation.

### Create a box plot for each portfolio


```python
# Using .boxplot to create a box plot to visually show risk
combined_returns_df.boxplot(figsize=(20,10));
```

### Calculate Standard Deviations

```python
# Calculating the daily standard deviations of all portfolios with .std()
combined_returns_df_std = combined_returns_df.std()
#Calling to view the standard deviation figures
combined_returns_df_std
```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculating the daily standard deviation of S&P TSX 60 with .std() seperately so it has seperate variable
sp_tsx_history_daily_std = combined_returns_df["S&P TSX 60 Returns"].std()
# Calling to view the standard deviation figure for S&P TSX 60 Returns
sp_tsx_history_daily_std
```

```python
# Determining which portfolios are riskier than the S&P TSX 60
# Want a boolean value to tell us which ones are more risky as 'True' and less riskier as 'False'
sp_tsx_std_risk = combined_returns_df_std > sp_tsx_history_daily_std
# Sorting the values so its easier to determine when we call it 
sp_tsx_std_risk.sort_values(ascending=False, inplace=True)
# Calling to confirm with the Boolean outputs
sp_tsx_std_risk
```

### Calculate the Annualized Standard Deviation

```python
# Calculating the annualized standard deviation with 252 trading days will indicate the Volatility
combined_annual_std_vol = combined_returns_df.std() * np.sqrt(252)
# Sorting the values in ascending order to make it easier to determine the volitility of the portfolios 
combined_annual_std_vol.sort_values(ascending=True, inplace=True)
# Calling to view the annual std deviation 
combined_annual_std_vol
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculating the rolling standard deviation for all portfolios using a 21-day window
combined_rol_std_21_day = combined_returns_df.rolling(window=21).std()
# Ploting the rolling standard deviation
combined_rol_std_21_day.plot(figsize=(20,10), title="Rolling Standard Deviation Of All Portfolios (21 Days)", ylabel="Rolling Standard Deviation");
```

### Calculate and plot the correlation

```python
# Calculating the correlation between all the portfolios
combined_returns_df_correlation = combined_returns_df.corr()
# Displaying the correlation matrix/heatmap with number annotation to make it easier to interpret  
sns.heatmap(combined_returns_df_correlation, vmin = -1, vmax = 1, annot = True);
```

### Calculate and Plot Beta for Tiger Global and the S&P 60 TSX

```python
# Calculating the covariance of a Tiger Global
tiger_global_rol_covariance = combined_returns_df["TIGER GLOBAL MANAGEMENT LLC"].rolling(window=60).cov(combined_returns_df["S&P TSX 60 Returns"])
# Calculating the variance of S&P TSX
sp_tsx_rol_variance = combined_returns_df["S&P TSX 60 Returns"].rolling(window=60).var()
# Calculating the beta for Tiger Global and S&P TSX
rol_beta_tiger_gloval_sp_tsx = tiger_global_rol_covariance / sp_tsx_rol_variance
# Ploting the beta trend for Tiger Global and S&P TSX
rol_beta_tiger_gloval_sp_tsx.plot(figsize=(20,10), title="Rolling Beta Trend For the S&P TSX 60 Returns and Tiger Global Management", ylabel="Beta Relation");
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Using `ewm` to calculate the average rolling window with hallife as 21 
combined_df_rol_ewm = combined_returns_df.ewm(halflife=21).mean()
# Plotting the 'ewm' result
combined_df_rol_ewm.plot(figsize=(20,10), title="Combined Dataframe of Portfolios | Expotentially Weighted Moving Average", ylabel="Returns");
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculating the annualized Sharpe Ratios for all portfolios at once
combined_df_sharpe_ratio = ((combined_returns_df.mean() * 252) / (combined_returns_df.std() * np.sqrt(252)))
# Sorting the values
combined_df_sharpe_ratio.sort_values(ascending=True, inplace=True)
# Calling to see the sharpe ratios 
combined_df_sharpe_ratio
```

```python
# Using a bar plot to view the sharpe ratios 
combined_df_sharpe_ratio.plot(kind='bar', figsize=(20,10), title="Sharpe Ratios", ylabel="Sharpe Ratio", xlabel="Portfolios");
```
